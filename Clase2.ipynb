# Tipos de datos string

```python
Nombre1 = "Sebastian"
Nombre2 =  'Gabriela'
print(Nombre1)
print(Nombre2)
```

```python
# concadenacion
print(Nombre1 + " " + Nombre2)
```

```python
'Sebastian' in globals().get('Nombre1', '')
```

```python
# indexado y segmentacion (slicing)
texto1 = "Hola mundo"
# []
# En texto1 cada letra tiene un indice, el primer caracter tiene el indice 0, el segundo el indice 1 y asi sucesivamente
texto1[0] # H
```

```python
#[inicio:fin:paso]
texto1[0:4] # Hola
```

```python
texto1[0:10:2] # Hlmno
```

```python
texto1[-1] # Hlmno
```

```python
type(texto1) # str
```

```python
.lower() # convertir a minusculas
.upper() # convertir a mayusculas
.title() # convertir a titulo
.capitalize() # convertir a capital
```

```python
nombre = "sebastian"
nombre.capitalize() # Sebastian
```

```python
# comandos find y replace
.find() # buscar un caracter o palabra en una cadena de texto
.replace() # reemplazar un caracter o palabra en una cadena de texto
```

```python
mensaje = "Apreciado sebastian, le informamos que su solicitud ha sido aprobada"
mensaje.replace("sebastian", "Gabriela") # Apreciado Gabriela, le informamos que su solicitud ha sido aprobada
# no olvidar actualizar la variable si se desea guardar el cambio
```

```python
mensaje.find("sebastian") # 10
```
